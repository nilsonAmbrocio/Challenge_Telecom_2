```python
import pandas as pd
```

```python
datos=pd.read_csv('/content/datos_tratados.csv')
datos
```

```python
datos.columns
```

```python
columnas_eliminar=['customerID', 'gender', 'MultipleLines','StreamingTV','StreamingMovies','Cuentas_Mensual']
datos=datos.drop(columns=columnas_eliminar)
datos
```

```python
for columna in datos.columns:
  print(columna)
  print(datos[columna].unique())
```

```python
import warnings
warnings.filterwarnings('ignore')#ignora advertencias
```

```python
#Reemplazar 0 y 1: Desercion,Partner,Dependents,PhoneService,Facturacion_electronica
datos['Desercion']=datos['Desercion'].replace({'No':0, 'Yes':1})
datos['Partner']=datos['Partner'].replace({'No':0, 'Yes':1})
datos['Dependents']=datos['Dependents'].replace({'No':0, 'Yes':1})
datos['PhoneService']=datos['PhoneService'].replace({'No':0, 'Yes':1})
datos['Facturacion_electronica']=datos['Facturacion_electronica'].replace({'No':0, 'Yes':1})
datos.sample(5)
```

```python
#Columnas categoricas: InternetService,OnlineSecurity,OnlineBackup,DeviceProtection,TechSupport,Contract,PaymentMethod

categoricas=['InternetService','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','Contract','PaymentMethod']
pd.get_dummies(data=datos, columns=categoricas, dtype=int).head()
```

```python
datos_codificados=pd.get_dummies(data=datos, columns=categoricas, dtype=int)
datos_codificados.sample(5)
```

```python
datos_codificados['Desercion'].value_counts(normalize=True) #Existe un desbalance en la variable Desercion
```

Correlacion

```python
datos_codificados.info()
```

```python
corr=datos_codificados.corr()
corr['Desercion'].sort_values(ascending=False)# mayor relacion con contrato mes a mes, antiguedad, seguridad online, soporte tecnico, servicio de internet de fibra optica
#contrato de 2 años, metodo de pago checke electronico, backup online, proteccion de dispositivos, etc.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
sns.boxplot(x='Desercion', y='Antiguedad', data=datos, ax=axes[0,0])
axes[0,0].set_title('Antiguedad vs Deserción')

sns.boxplot(x='Desercion', y='Cuentas_Total', data=datos, ax=axes[0,1])
axes[0,1].set_title('Cuentas Total vs Deserción')

sns.boxplot(x='Desercion', y='Antiguedad', data=datos_codificados, ax=axes[1,0])
axes[1,0].set_title('Antiguedad vs Deserción Codificados')

sns.boxplot(x='Desercion', y='Cuentas_Total', data=datos_codificados, ax=axes[1,1])
axes[1,1].set_title('Cuentas Total vs Deserción Codificados')

plt.show()
```

Oversampling

```python
from imblearn.over_sampling import SMOTE
```

```python
x = datos_codificados.drop('Desercion', axis=1)  # Características
y = datos_codificados['Desercion']                # Variable objetivo

oversampling=SMOTE()
x_balanceada, y_balanceada=oversampling.fit_resample(x,y)
```

```python
y_balanceada.value_counts(normalize=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(x_balanceada,y_balanceada, test_size=0.3, stratify=y_balanceada, random_state=5)
```

```python
from sklearn.model_selection import KFold, cross_validate
```

```python
from sklearn.tree import DecisionTreeClassifier

modelo=DecisionTreeClassifier(max_depth=10)
kf=KFold(n_splits=5, shuffle=True, random_state=5)
cv_resultados_recall=cross_validate(modelo, x_train, y_train, cv=kf, scoring='recall')
cv_resultados_recall
```

```python
def intervalo_confianza(resultados):
  promedio=resultados['test_score'].mean()
  desvio_estandar=resultados['test_score'].std()
  return f'El intervalo de confianza de la exactitud del modelo es [{promedio-2*desvio_estandar}, {promedio+2*desvio_estandar}] y el promedio es {promedio}'
```

```python
modelo=DecisionTreeClassifier(max_depth=10)
kf=KFold(n_splits=5, shuffle=True, random_state=5)
cv_resultados_accuracy=cross_validate(modelo, x_train, y_train, cv=kf, scoring='accuracy')
```

```python
modelo=DecisionTreeClassifier(max_depth=10)
kf=KFold(n_splits=5, shuffle=True, random_state=5)
cv_resultados_precision=cross_validate(modelo, x_train, y_train, cv=kf, scoring='precision')
```

```python
modelo=DecisionTreeClassifier(max_depth=10)
kf=KFold(n_splits=5, shuffle=True, random_state=5)
cv_resultados_f1=cross_validate(modelo, x_train, y_train, cv=kf, scoring='f1')
```

```python
print(f'ACCURACY:{intervalo_confianza(cv_resultados_accuracy)}')#accuracy
print(f'RECALL: {intervalo_confianza(cv_resultados_recall)}')#recall
print(f'PRECISION:{intervalo_confianza(cv_resultados_precision)}')#precision
print(f'F1-SCORE:{intervalo_confianza(cv_resultados_f1)}')#f1-score
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import ConfusionMatrixDisplay

modelo.fit(x_train, y_train)
y_previsto=modelo.predict(x_test)

print(classification_report(y_test, y_previsto))
ConfusionMatrixDisplay.from_predictions(y_test, y_previsto);
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

print(f'la presicion del modelo es {precision_score(y_test,y_previsto)}')
print(f'la sensibilidad (Recall) del modelo es {recall_score(y_test,y_previsto)}')
print(f'la exactitud (accuracy) del modelo es {accuracy_score(y_test,y_previsto)}')
print(f'el F1-score del modelo es {f1_score(y_test,y_previsto)}')
```

```python
importances=modelo.feature_importances_
```

```python
feature_importances=pd.DataFrame({'Features':x.columns, 'Importances':(importances*100).round(2)}).sort_values('Importances', ascending=False)
feature_importances
```

Basandonos en el modelo de clasificacion DecisionTreeClassifier, los factores que mas influyen en la Desercion o Cancelacion de clientes son:



1.   Los clientes que tienen contratos mensuales
2.   Los clientes que tienen contratos de 1 año
3.   Los clientes que gastan menos





# MODELO KNN

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
normalizacion=MinMaxScaler()
x_train_normalizada=normalizacion.fit_transform(x_train)
pd.DataFrame(x_train_normalizada)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier()
knn.fit(x_train_normalizada, y_train)
```

```python
x_test_normalizada=normalizacion.transform(x_test)
```

```python
knn.score(x_test_normalizada, y_test)
```

```python
kf=KFold(n_splits=5, shuffle=True, random_state=5)
cv_resultados_recall_knn=cross_validate(knn, x_train, y_train, cv=kf, scoring='recall')
print(f'RECALL: {intervalo_confianza(cv_resultados_recall_knn)}')#recall
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import ConfusionMatrixDisplay

y_previsto=knn.predict(x_test_normalizada)

print(classification_report(y_test, y_previsto))
ConfusionMatrixDisplay.from_predictions(y_test, y_previsto);
```

```python
# Importancia por permutación
from sklearn.inspection import permutation_importance

resultado = permutation_importance(knn, x_test_normalizada, y_test, n_repeats=10, random_state=42)

importancias = pd.DataFrame({
    "Variable": x_test.columns,
    "Importancia": (resultado.importances_mean*100).round(2)
}).sort_values(by="Importancia", ascending=False)

print(importancias)
```

Basandonos en el modelo de clasificacion KNearestNeighbours, los factores que mas influyen en la Desercion o Cancelacion de clientes son:

1.   Los clientes que tienen menor antiguedad
2.   Los clientes que gastan menos
